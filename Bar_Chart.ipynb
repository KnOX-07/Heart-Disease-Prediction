```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
from matplotlib import style
```

```python
dataset = pd.read_excel('/content/accuracy.xlsx')
#To display the content
dataset.head()
```

```python
algorithm_name = dataset['Algorithm Name ']
accuracy_score = dataset['Accuracy Score']
```

```python
fig = plt.figure(figsize = (8,5))
# creating the bar plot
plt.bar(algorithm_name, accuracy_score, color ='maroon',
        width = 0.7)
plt.xlabel('Algorithm Name ', fontweight ='bold', fontsize = 10)
plt.ylabel('Accuracy Score', fontweight ='bold', fontsize = 10)
plt.title('Heart Disease Prediction using ML algorithms')
plt.show()
```

```python
fig = plt.figure(figsize = (8,5))
# creating the bar plot
plt.barh(algorithm_name, accuracy_score, color ='blue')
plt.xlabel('Accuracy Score', fontweight ='bold', fontsize = 10)
plt.ylabel('Algorithm Name ', fontweight ='bold', fontsize = 10)
plt.title('Heart Disease Prediction using ML algorithms')
plt.show()
```
